```python
import tkinter as tk
from PIL import ImageTk, Image

# Create a tkinter window and maximize it
root = tk.Tk()
root.state('zoomed')

# Create a list of images
image_list = ['img001-11.jpg', 'img001-12.jpg', 'img001-13.jpg']

# Create a list of tk.Label widgets and PhotoImage objects
label_list = []
photo_list = []

# Loop through the image list and create a Label widget and a PhotoImage object for each image
for i in range(len(image_list)):
    img = Image.open(image_list[i])
    photo = ImageTk.PhotoImage(img)
    label = tk.Label(root, image=photo)
    label_list.append(label)
    photo_list.append(photo)

# Loop through the label_list and use the grid method to place each label widget on the window
for i in range(len(label_list)):
    label_list[i].grid(row=i//2, column=i%2, padx=5, pady=5, sticky="nsew")

# Configure grid to make images expandable
for i in range(len(image_list)):
    root.grid_columnconfigure(i % 2, weight=1)
    root.grid_rowconfigure(i // 2, weight=1)

# Start the mainloop
root.mainloop()
```
